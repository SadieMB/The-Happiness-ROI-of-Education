```python
#filtered rows to show results for total population (not the inequality measures)
happy_df = Happiness_df[Happiness_df['INEQUALITY'] == 'TOT']
happy_df.dtypes
```

```python
#re-organized columns and rows with pivot to see the values of each indicator for each country.
pivot=pd.pivot_table(data=happy_df, values='Value', index='Country', columns='Indicator',aggfunc='mean') 
print(pivot)
```

```python
#Checking data types (again)
pivot.dtypes

```

```python
Max_df = pd.DataFrame(pivot.max(axis=0),columns=["Max Value"])
Max_df.columns = Max_df.columns.to_flat_index()

Min_df = pd.DataFrame(pivot.min(axis=0),columns=["Min Value"])
Min_df.columns = Min_df.columns.to_flat_index()

Influence_df = pd.concat([Max_df,Min_df], axis=1, join='inner').reset_index()
print(Influence_df)
```

```python
Inf_df2 = Influence_df.assign(LSofMin = Influence_df['Indicator'].apply(lambda x:(pivot.loc[pivot[x].idxmin()]['Life satisfaction'])))
print(Inf_df2)
```

```python
Inf_df3=Inf_df2.assign(LSofMax = Influence_df['Indicator'].apply(lambda x:(pivot.loc[pivot[x].idxmax()]['Life satisfaction'])))
print(Inf_df3)
```

```python
Inf_df3.dtypes
```

```python
Inf_df4 = Inf_df3.assign(LS_diff = abs(Inf_df3['LSofMin']-Inf_df3['LSofMax']))
print(Inf_df4)
```

```python
Inf_df4.to_csv('Influence.csv',index=True)
```
